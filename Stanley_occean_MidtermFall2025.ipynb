```python
import re

def count_vowels(s: str) -> int:
    return sum(ch.lower() in "aeiou" for ch in s)

def is_palindrome(s: str) -> bool:
    t = re.sub(r"[^A-Za-z0-9]", "", s).lower()
    return t == t[::-1]

def mask_email(email: str) -> str:
    local, domain = email.split("@")
    if len(local) <= 2:
        masked_local = local[0] + "*"
    else:
        masked_local = local[0] + "*"*(len(local)-2) + local[-1]
    return masked_local + "@" + domain

count_vowels("Parfait Time"), is_palindrome("Never odd or even"), mask_email("star.user@coffee.com")
```

```python
import random, string

print("Welcome to the PyPassword Generator!")
n_letters = int(input("How many letters would you like in your password?\n"))
n_symbols = int(input("How many symbols would you like?\n"))
n_numbers = int(input("How many numbers would you like?\n"))

letters = list(string.ascii_letters)
symbols = list("!@#$%^&*()-_=+[]{};:,.?/\\|~")
numbers = list(string.digits)

chars = (
    random.choices(letters, k=n_letters)
    + random.choices(symbols, k=n_symbols)
    + random.choices(numbers, k=n_numbers)
)
random.shuffle(chars)
password = "".join(chars)
print(f"Your password is: {password}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"
resp = requests.get(url, timeout=30)
resp.raise_for_status()

soup = BeautifulSoup(resp.text, "html.parser")

tables = pd.read_html(resp.text)
df_fdic = None
for t in tables:
    cols = [c.lower() for c in t.columns.astype(str)]
    if ("bank name" in cols and "city" in cols and "state" in cols):
        df_fdic = t
        break

if df_fdic is None:
    table = soup.find("table")
    headers = [th.get_text(strip=True) for th in table.find_all("th")]
    rows = []
    for tr in table.find_all("tr")[1:]:
        tds = [td.get_text(" ", strip=True) for td in tr.find_all("td")]
        if tds:
            rows.append(tds)
    df_fdic = pd.DataFrame(rows, columns=headers)

df_fdic.columns = [c.strip() for c in df_fdic.columns]
if "Bank Name" in df_fdic.columns:
    links = []
    table = soup.find("table")
    for a in table.select("tbody tr td:nth-child(1) a"):
        links.append(a.get("href"))
    if links and len(links) == len(df_fdic):
        df_fdic.insert(1, "Link", links)

df_fdic.head(10)
```
